```python
from sklearn.model_selection import train_test_split
```

```python
features =emiss.drop('CO2 Emissions(g/km)',axis=1)
```

```python
target = emiss["CO2 Emissions(g/km)"]
```

```python
X_train,X_test,y_train,y_test = train_test_split(pd.get_dummies(features),target,test_size=0.2)
```

```python
X_train.columns
```

```python
from sklearn.naive_bayes import GaussianNB  
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error,accuracy_score
```

```python
Gass=GaussianNB ()
Gass.fit(X_train,y_train)
y_pred=Gass.predict(X_test)

acc=r2_score(y_test,y_pred)

Acu=accuracy_score(y_test,y_pred)
mse=mean_squared_error(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)
print("Accuracy:",Acu)
print("r2_score:",acc)
print("Mean Squared Error:",mse)
print("Mean Absolute Error:",mae)
```

```python
actual_value=y_test
prediction=y_pred
print("Actual Value:",actual_value[:5])
print("Predicted Value",prediction[:5])
```

```python
sns.regplot(x=actual_value,y=prediction)
plt.xlabel("Actual Value")
plt.ylabel("Predicted Value")
plt.title("Actual Value Vs Predicted Value")
```
